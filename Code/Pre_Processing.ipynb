```python
import numpy as np
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Read Combined ETF.csv
trail_df = pd.read_csv(
    Path("../Data/ETF CSV files/Combined ETF.csv"), index_col="ETF Names",
)

trail_df.head()
```

```python
# Change NaN in partial years to 0
trail_df['30-Jun-17'] = trail_df['30-Jun-17'].fillna(0)
trail_df['31-Jul-17'] = trail_df['31-Jul-17'].fillna(0)

# Trim DF
ETF_2017_2022_df = trail_df.loc["iShares Asia 50 ETF":"iShares Yield Plus ETF", "30-Jun-17":"30-Jun-22"]
ETF_2017_2022_df.head()

# Drop ETF rows with majority NaN //// get this corrected
ETF_2017_2022_df.drop(ETF_2017_2022_df[ETF_2017_2022_df[ETF_2017_2022_df.columns].isnull().sum(axis=1) > 3].index, inplace = True)
ETF_2017_2022_df.head(36)
```

```python
# Calculate STD
STD_2017_2022 = np.transpose(ETF_2017_2022_df).std()

# Calculate Mean
Mean_2017_2022 = np.transpose(ETF_2017_2022_df).mean()

# Add STD and Mean Column
df_1 = ETF_2017_2022_df.copy()
df_1["STD_2017_2022"] = STD_2017_2022
df_1["Mean_2017_2022"] = Mean_2017_2022

# Trim DF to desired columns
df_1a = df_1[["STD_2017_2022", "Mean_2017_2022"]].copy()

df_1a.head()
```
