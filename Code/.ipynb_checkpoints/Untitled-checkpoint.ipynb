```python
import numpy as np
import pandas as pd
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")
```

```python
# Read CSVs 
Asia_50_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Asia-50-ETF_fund.csv"), index_col="Date")
Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Core-MSCI-World-ex-Australia-ESG-Leaders-AUD-Hedged_fund.csv"), index_col="Date")
Core_MSCI_World_ex_Australia_ESG_Leaders_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Core-MSCI-World-ex-Australia-ESG-Leaders-ETF_fund.csv"), index_col="Date")
Core_SPASX_200_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Core-SPASX-200-ETF_fund.csv"), index_col="Date")
Edge_MSCI_Australia_Minimum_Volatility_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Edge-MSCI-Australia-Minimum-Volatility-ETF_fund.csv"), index_col="Date")
Edge_MSCI_Australia_Multifactor_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Edge-MSCI-Australia-Multifactor-ETF_fund.csv"), index_col="Date")
Edge_MSCI_World_Multifactor_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Edge-MSCI-World-Multifactor-ETF_fund.csv"), index_col="Date")
Global_100_AUD_Hedged_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Global-100-AUD-Hedged-ETF_fund.csv"), index_col="Date")
Global_100_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Global-100-ETF_fund.csv"), index_col="Date")
Global_Consumer_Staples_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Global-Consumer-Staples-ETF_fund.csv"), index_col="Date")
Global_Healthcare_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-Global-Healthcare-ETF_fund.csv"), index_col="Date")
JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-JP-Morgan-USD-Emerging-Markets-Bond-AUD-Hedged-E_fund.csv"), index_col="Date")
MSCI_Emerging_Markets_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-MSCI-Emerging-Markets-ETF_fund.csv"), index_col="Date")
SP_500_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-SP-500-ETF_fund.csv"), index_col="Date")
SP_Mid_Cap_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-SP-Mid-Cap-ETF_fund.csv"), index_col="Date")
SPASX_20_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-SPASX-20-ETF_fund.csv"), index_col="Date")
SPASX_Dividend_Opportunities_df = pd.read_csv(Path("../Data/ETF CSV files/Singular ETF data/iShares-SPASX-Dividend-Opportunities-ETF_fund.csv"), index_col="Date")
```

```python
# Drop Volume in dfs
vless_Asia_50_df = Asia_50_df.drop(columns=["Volume"])
vless_Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_df = Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_df.drop(columns=["Volume"])
vless_Core_MSCI_World_ex_Australia_ESG_Leaders_df = Core_MSCI_World_ex_Australia_ESG_Leaders_df.drop(columns=["Volume"])
vless_Core_SPASX_200_df = Core_SPASX_200_df.drop(columns=["Volume"])
vless_Edge_MSCI_Australia_Minimum_Volatility_df = Edge_MSCI_Australia_Minimum_Volatility_df.drop(columns=["Volume"])
vless_Edge_MSCI_Australia_Multifactor_df = Edge_MSCI_Australia_Multifactor_df.drop(columns=["Volume"])
vless_Edge_MSCI_World_Multifactor_df = Edge_MSCI_World_Multifactor_df.drop(columns=["Volume"])
vless_Global_100_AUD_Hedged_df = Global_100_AUD_Hedged_df.drop(columns=["Volume"])
vless_Global_100_df = Global_100_df.drop(columns=["Volume"])
vless_Global_Consumer_Staples_df = Global_Consumer_Staples_df.drop(columns=["Volume"])
vless_Global_Healthcare_df = Global_Healthcare_df.drop(columns=["Volume"])
vless_JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_df = JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_df.drop(columns=["Volume"])
vless_MSCI_Emerging_Markets_df = MSCI_Emerging_Markets_df.drop(columns=["Volume"])
vless_SP_500_df = SP_500_df.drop(columns=["Volume"])
vless_SP_Mid_Cap_df = SP_Mid_Cap_df.drop(columns=["Volume"])
vless_SPASX_20_df = SPASX_20_df.drop(columns=["Volume"])
vless_SPASX_Dividend_Opportunities_df = SPASX_Dividend_Opportunities_df.drop(columns=["Volume"])
```

```python
# Rename Costs columns to ETF name Costs
vless_Asia_50_df.rename(columns = {"Costs":"Asia_50_Costs"}, inplace = True)
vless_Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_df.rename(columns = {"Costs":"Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_Costs"}, inplace = True)
vless_Core_MSCI_World_ex_Australia_ESG_Leaders_df.rename(columns = {"Costs":"Core_MSCI_World_ex_Australia_ESG_Leaders_Costs"}, inplace = True)
vless_Core_SPASX_200_df.rename(columns = {"Costs":"Core_SPASX_200_Costs"}, inplace = True)
vless_Edge_MSCI_Australia_Minimum_Volatility_df.rename(columns = {"Costs":"Edge_MSCI_Australia_Minimum_Volatility_Costs"}, inplace = True)
vless_Edge_MSCI_Australia_Multifactor_df.rename(columns = {"Costs":"Edge_MSCI_Australia_Multifactor_Costs"}, inplace = True)
vless_Edge_MSCI_World_Multifactor_df.rename(columns = {"Costs":"Edge_MSCI_World_Multifactor_Costs"}, inplace = True)
vless_Global_100_AUD_Hedged_df.rename(columns = {"Costs":"Global_100_AUD_Hedged_Costs"}, inplace = True)
vless_Global_100_df.rename(columns = {"Costs":"Global_100_Costs"}, inplace = True)
vless_Global_Consumer_Staples_df.rename(columns = {"Costs":"Global_Consumer_Staples_Costs"}, inplace = True)
vless_Global_Healthcare_df.rename(columns = {"Costs":"Global_Healthcare_Costs"}, inplace = True)
vless_JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_df.rename(columns = {"Costs":"JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_Costs"}, inplace = True)
vless_MSCI_Emerging_Markets_df.rename(columns = {"Costs":"MSCI_Emerging_Markets_Costs"}, inplace = True)
vless_SP_500_df.rename(columns = {"Costs":"SP_500_Costs"}, inplace = True)
vless_SP_Mid_Cap_df.rename(columns = {"Costs":"SP_Mid_Cap_Costs"}, inplace = True)
vless_SPASX_20_df.rename(columns = {"Costs":"SPASX_20_Costs"}, inplace = True)
vless_SPASX_Dividend_Opportunities_df.rename(columns = {"Costs":"SPASX_Dividend_Opportunities_Costs"}, inplace = True)
```

```python
# Concat the Volumeless dataframes together
vless_combined_df = pd.concat([vless_Asia_50_df,
                         vless_Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_df,
                         vless_Core_MSCI_World_ex_Australia_ESG_Leaders_df,
                         vless_Core_SPASX_200_df,
                         vless_Edge_MSCI_Australia_Minimum_Volatility_df,
                         vless_Edge_MSCI_Australia_Multifactor_df,
                         vless_Edge_MSCI_World_Multifactor_df,
                         vless_Global_100_AUD_Hedged_df,
                         vless_Global_100_df,
                         vless_Global_Consumer_Staples_df,
                         vless_Global_Healthcare_df,
                         vless_JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_df,
                         vless_MSCI_Emerging_Markets_df,
                         vless_SP_500_df,
                         vless_SP_Mid_Cap_df,
                         vless_SPASX_20_df,
                         vless_SPASX_Dividend_Opportunities_df], axis=1).rename_axis("Dates")

# Check vless_combined_df
vless_combined_df.head()
```

```python
# Replace NaN with 0
vless_combined_df.fillna(0, inplace=True)

# Check for NaNs
vless_combined_df.isnull().values.any()
```

```python
# Calculate mean of all the Costs
vless_combined_df["Total_Costs"] = vless_combined_df.mean(axis=1)
vless_combined_df.head()
```

```python
# Drop Costs in dfs
cless_Asia_50_df = Asia_50_df.drop(columns=["Costs"])
cless_Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_df = Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_df.drop(columns=["Costs"])
cless_Core_MSCI_World_ex_Australia_ESG_Leaders_df = Core_MSCI_World_ex_Australia_ESG_Leaders_df.drop(columns=["Costs"])
cless_Core_SPASX_200_df = Core_SPASX_200_df.drop(columns=["Costs"])
cless_Edge_MSCI_Australia_Minimum_Volatility_df = Edge_MSCI_Australia_Minimum_Volatility_df.drop(columns=["Costs"])
cless_Edge_MSCI_Australia_Multifactor_df = Edge_MSCI_Australia_Multifactor_df.drop(columns=["Costs"])
cless_Edge_MSCI_World_Multifactor_df = Edge_MSCI_World_Multifactor_df.drop(columns=["Costs"])
cless_Global_100_AUD_Hedged_df = Global_100_AUD_Hedged_df.drop(columns=["Costs"])
cless_Global_100_df = Global_100_df.drop(columns=["Costs"])
cless_Global_Consumer_Staples_df = Global_Consumer_Staples_df.drop(columns=["Costs"])
cless_Global_Healthcare_df = Global_Healthcare_df.drop(columns=["Costs"])
cless_JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_df = JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_df.drop(columns=["Costs"])
cless_MSCI_Emerging_Markets_df = MSCI_Emerging_Markets_df.drop(columns=["Costs"])
cless_SP_500_df = SP_500_df.drop(columns=["Costs"])
cless_SP_Mid_Cap_df = SP_Mid_Cap_df.drop(columns=["Costs"])
cless_SPASX_20_df = SPASX_20_df.drop(columns=["Costs"])
cless_SPASX_Dividend_Opportunities_df = SPASX_Dividend_Opportunities_df.drop(columns=["Costs"])
```

```python
# Rename Volume columns to ETF name Volume
cless_Asia_50_df.rename(columns = {"Volume":"Asia_50_Volume"}, inplace = True)
cless_Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_df.rename(columns = {"Volume":"Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_Volume"}, inplace = True)
cless_Core_MSCI_World_ex_Australia_ESG_Leaders_df.rename(columns = {"Volume":"Core_MSCI_World_ex_Australia_ESG_Leaders_Volume"}, inplace = True)
cless_Core_SPASX_200_df.rename(columns = {"Volume":"Core_SPASX_200_Volume"}, inplace = True)
cless_Edge_MSCI_Australia_Minimum_Volatility_df.rename(columns = {"Volume":"Edge_MSCI_Australia_Minimum_Volatility_Volume"}, inplace = True)
cless_Edge_MSCI_Australia_Multifactor_df.rename(columns = {"Volume":"Edge_MSCI_Australia_Multifactor_Volume"}, inplace = True)
cless_Edge_MSCI_World_Multifactor_df.rename(columns = {"Volume":"Edge_MSCI_World_Multifactor_Volume"}, inplace = True)
cless_Global_100_AUD_Hedged_df.rename(columns = {"Volume":"Global_100_AUD_Hedged_Volume"}, inplace = True)
cless_Global_100_df.rename(columns = {"Volume":"Global_100_Volume"}, inplace = True)
cless_Global_Consumer_Staples_df.rename(columns = {"Volume":"Global_Consumer_Staples_Volume"}, inplace = True)
cless_Global_Healthcare_df.rename(columns = {"Volume":"Global_Healthcare_Volume"}, inplace = True)
cless_JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_df.rename(columns = {"Volume":"JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_Volume"}, inplace = True)
cless_MSCI_Emerging_Markets_df.rename(columns = {"Volume":"MSCI_Emerging_Markets_Volume"}, inplace = True)
cless_SP_500_df.rename(columns = {"Volume":"SP_500_Volume"}, inplace = True)
cless_SP_Mid_Cap_df.rename(columns = {"Volume":"SP_Mid_Cap_Volume"}, inplace = True)
cless_SPASX_20_df.rename(columns = {"Volume":"SPASX_20_Volume"}, inplace = True)
cless_SPASX_Dividend_Opportunities_df.rename(columns = {"Volume":"SPASX_Dividend_Opportunities_Volume"}, inplace = True)
```

```python
# Concat the Costless dataframes together
cless_combined_df = pd.concat([cless_Asia_50_df,
                         cless_Core_MSCI_World_ex_Australia_ESG_Leaders_AUD_Hedged_fund_df,
                         cless_Core_MSCI_World_ex_Australia_ESG_Leaders_df,
                         cless_Core_SPASX_200_df,
                         cless_Edge_MSCI_Australia_Minimum_Volatility_df,
                         cless_Edge_MSCI_Australia_Multifactor_df,
                         cless_Edge_MSCI_World_Multifactor_df,
                         cless_Global_100_AUD_Hedged_df,
                         cless_Global_100_df,
                         cless_Global_Consumer_Staples_df,
                         cless_Global_Healthcare_df,
                         cless_JP_Morgan_USD_Emerging_Markets_Bond_AUD_Hedged_E_df,
                         cless_MSCI_Emerging_Markets_df,
                         cless_SP_500_df,
                         cless_SP_Mid_Cap_df,
                         cless_SPASX_20_df,
                         cless_SPASX_Dividend_Opportunities_df], axis=1).rename_axis("Dates")

# Check vless_combined_df
cless_combined_df.head()
```
